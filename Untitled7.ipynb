#Label Encoder-Converts categorical data like labels or strings into numerical format while One hot encoder converts the data only in 0 and 1 while this gives a range of number associated with data

```python
import pandas as pd
```

#Label Encoder

```python
df=pd.DataFrame({"Name":["Daksh","Arima","Ashvika","Mridu","Rishi","Vaishnavi","Arima"]})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
le.fit_transform(df["Name"])
```

```python
df["en_name"]=le.fit_transform(df["Name"])
```

```python
df
```

```python
new_data=df
new_data.to_csv("Label_Encoder.csv",index=False)
```
